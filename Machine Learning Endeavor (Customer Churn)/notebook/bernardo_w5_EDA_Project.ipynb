## A. Univariate analysis including histogram and box plot with written description for the following variables:

### I. Customer Age

````python
#### i. 'Customer_Age' Description
pdf['Customer_Age'].describe()
````

````python
#### i. 'Customer_Age' box plot
pdf['Customer_Age'].plot(kind='box')
````

The plot signifies the distribution of customer ages, the range of the data is from 26 to 70. The box shows that most customers ages fall between 41 to 52 with the mean and median to be around 46 years old denoted by the red line, signifying that most customers could be around 46 years old. Based on the overall outlook, this could mean that the bank is able to cater most of the age groups.

````python
#### i. 'Customer_Age' histogram
pdf['Customer_Age'].plot(kind='hist')
````

The histogram signifies that the most common age group falls around 40-50 years old. With the bars for older and younger age group being short, it simply means that there are fewer customers to those age range. Based on the overall outlook, this could mean that the bank's service offers are highly attractive  to ages 40-50, could this mean that they are lacking on attractive service offerings on younger or even older age range? Or is it exactly the target market they wanted?

### II. Dependent Count / Supplementary Cards

````python
#### i. 'Dependent_count' Description
pdf['Dependent_count'].describe()
````

````python
#### ii. 'Dependent_count' box plot
pdf['Dependent_count'].plot(kind='box')
````

The box plot represents the customers dependent count or the number of supplementary cards they possess. The range of the data is from 0 to 5. The box shows that most customers have between 1 to 3 supplementary cards with a median found to be 2. The histogram shows that most customers have 2-3 supplmentary cards. There is a small number of those without supplementary cards and those who have 5. While there are a significant amount of customers having 1 or 4 dependencies. Based on the overall outlook, this could suggest that customers prefer having 2-3 cards at most for it is sufficient enough for whatever they need.

## C. Bivariate analysis including descriptive statistics and bar plot with written description for the following pair of variables:

### I. 'Customer_Age' vs 'Attrition_Flag'

````python
#### 1. 'Customer_Age' vs 'Attrition_Flag' mean
pdf.groupby('Attrition_Flag').mean()['Customer_Age']
````

````python
#### i. 'Customer_Age' vs 'Attrition_Flag' bar plot
pdf.groupby('Attrition_Flag').mean()['Customer_Age'].plot(kind='bar')
````

The bar plot shows the mean age of customers in relation to their attrition status. It shows that customers who have stayed have higher mean age 46.2, pretty close to those who churned who has 47 average age. This suggests that younger customers are slightly more likely to stay if we are talking about the difference, but in essence  average age of customers who churn is roughly the same as those who stay.
